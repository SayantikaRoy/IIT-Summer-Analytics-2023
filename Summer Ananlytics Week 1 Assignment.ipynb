```python
import pandas as pd
df = pd.read_csv("Documents/Week 1 Assignment/imdb_top_1000.csv")

df['Runtime'] = pd.to_numeric(df['Runtime'].str.replace(' min', ''))

max_runtime_row = df['Runtime'].idxmax()

max_runtime_genre = df.loc[max_runtime_row, 'Genre']
max_runtime_movie = df.loc[max_runtime_row, 'Series_Title']

print(f"The movie with the highest runtime in the genre '{max_runtime_genre}' is '{max_runtime_movie}'")
print(f"The genre with the highest runtime is: {max_runtime_genre}")
```

5. How many movies are released after year 2000 and have IMDB rating > 8.5 ?

```python
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')

filtered_movies = df[(df['Released_Year'] > 2000) & (df['IMDB_Rating'] > 8.5)]

num_movies = len(filtered_movies)

print("Number of movies released after 2000 with IMDb rating > 8.5:", num_movies)
```

6. Make a countplot of Movies directed by all directors.

```python
plt.figure(figsize=(10, 6))  

top_directors = df['Director'].value_counts().nlargest(5).index

sns.countplot(data=df[df['Director'].isin(top_directors)], x='Director')

plt.xticks(rotation=90)  
plt.xlabel('Director')  
plt.ylabel('Count')  
plt.title('Number of Movies Directed by Top 5 Directors')  

plt.show()
```

7. Make a plot to analyze distribution of IMDB rating?

```python
sns.displot(df['IMDB_Rating'], bins=10, kde=False, rug=True)
plt.xlabel('IMDB_Rating')
plt.ylabel('Frequency')
plt.title('Distribution of IMDB Rating')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you already have the 'df' dataframe with the 'Genre' and 'IMDB_Rating' columns

# Calculate the average IMDb rating for each genre
average_ratings = df.groupby('Genre')['IMDB_Rating'].mean().reset_index()

# Sort the genres based on average IMDb rating in descending order
sorted_genres = average_ratings.sort_values(by='IMDB_Rating', ascending=False)

# Create the plot
plt.figure(figsize=(20, 40))
sns.boxplot(x='IMDB_Rating', y='Genre', data=sorted_genres)
plt.xlabel('Average IMDb Rating')
plt.ylabel('Genre')
plt.title('Average IMDb Rating by Genre')
plt.show()
```

10. Of which Genre movie has the highest Runtime?

```python
df = pd.read_csv("Documents/Week 1 Assignment/imdb_top_1000.csv")
df['Runtime'] = pd.to_numeric(df['Runtime'].str.replace(' min', ''), errors='coerce')
max_runtime_row = df['Runtime'].idxmax()
max_runtime_genre = df.loc[max_runtime_row, 'Genre']
max_runtime_movie = df.loc[max_runtime_row, 'Series_Title']
max_runtime_runtime = df.loc[max_runtime_row, 'Runtime']
print(f"The movie with the highest runtime in the genre '{max_runtime_genre}' is '{max_runtime_movie}'")
print(f"The genre with the highest runtime is: {max_runtime_genre}")
print(f"The runtime of the movie is: {max_runtime_runtime}")
df_hr=df.sort_values(by=['Runtime'], ascending=False).head(500)
# print("Scatter Plot:  ")
# plt.scatter(df["Runtime"], df["Genre"])
# plt.show()
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(1,1,1)
ax.bar(df_hr["Genre"][:60],df_hr["Runtime"][:60])
plt.xticks(rotation=90)
plt.show()
```
